# Data preprocess

```python
df=dataset.copy()
```

```python
#month
#unique month
print(dataset.Month.unique())
df.loc[df['Month']=='Feb', 'Month'] = 1
df.loc[df['Month']=='Mar', 'Month'] = 2
df.loc[df['Month']=='May', 'Month'] = 3
df.loc[df['Month']=='Oct', 'Month'] = 4
df.loc[df['Month']=='June', 'Month'] = 5
df.loc[df['Month']=='Jul', 'Month'] = 6
df.loc[df['Month']=='Aug', 'Month'] = 7
df.loc[df['Month']=='Nov', 'Month'] = 8
df.loc[df['Month']=='Sep', 'Month'] = 9
df.loc[df['Month']=='Dec', 'Month'] = 0
print(df.Month.unique())
```

```python
#visitor type
#visitor type unique
print(dataset.VisitorType.unique())
df.loc[df['VisitorType']=='Returning_Visitor', 'VisitorType'] = 1
df.loc[df['VisitorType']=='New_Visitor', 'VisitorType'] = 2
df.loc[df['VisitorType']=='Other', 'VisitorType'] = 3
print(df.VisitorType.unique())
```

```python
#weekend
#weekend unique
print(dataset.Weekend.unique())
df['Weekend'] = df['Weekend'].astype(int)
print(df.Weekend.unique())
```

```python
x = df.drop('Revenue',1)
y = df['Revenue']
```

# K-Nearest  Neigbours,MLP,Naive Bayes and OLS

```python
import statsmodels.api as sm
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
```

```python
cols = list(x.columns)
pmax = 1
accuracy_knn=[]
accuracy_mlp=[]
accuracy_nb=[]
dimension=[]

while (len(cols)>0):
    features=cols
    z = df[features]
    
    knn = KNeighborsClassifier(n_neighbors=5)
    mlp = MLPClassifier(max_iter=1000,random_state=1)
    gnb = GaussianNB()
    
    strat_k_fold=StratifiedKFold(n_splits=10, shuffle=True, random_state=21)
    scores_knn = cross_val_score(knn, z, y, cv=strat_k_fold)
    scores_mlp = cross_val_score(mlp, z, y, cv=strat_k_fold)
    scores_nb = cross_val_score(gnb, z, y, cv=strat_k_fold)
    
    
    scores_knn = (scores_knn.mean() * 100)
    scores_mlp = (scores_mlp.mean() * 100)
    scores_nb = (scores_nb.mean() * 100)
    
    print("K-Nearest Neighbour Accuracy :", round(scores_knn, 2), "%" )
    print("Multi-Layer Perceptron Accuracy :", round(scores_mlp, 2), "%" )
    print("Naive Bayes Accuracy :", round(scores_nb, 2), "%" )
    
    accuracy_knn.append(round(scores_knn,2))
    accuracy_mlp.append(round(scores_mlp,2))
    accuracy_nb.append(round(scores_nb,2))
    dimension.append(len(features))
    
    p= []
    x_1 = x[cols]
    x_1 = sm.add_constant(x_1.values)
    model = sm.OLS(y,x_1).fit()
    p = pd.Series(model.pvalues.values[1:],index = cols)      
    pmax = max(p)
    feature_with_p_max = p.idxmax()
    if(pmax>0.05):
        
        #print(=features)
        print("\n\n----Removing feature (",feature_with_p_max,")-----")
        cols.remove(feature_with_p_max)
        features = cols
        
    else:
        break
```

```python
df_knn = pd.DataFrame(list(zip(dimension, accuracy_knn)), 
               columns =['Dimension', 'Accuracy(%)'])

df_mlp = pd.DataFrame(list(zip(dimension, accuracy_mlp)), 
               columns =['Dimension', 'Accuracy(%)']) 

df_nb = pd.DataFrame(list(zip(dimension, accuracy_nb)), 
               columns =['Dimension', 'Accuracy(%)']) 
```

# Visualizing Accuracy 

```python
import plotly.graph_objs as go
import plotly.io as pio
from IPython.display import Image
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=df_knn['Dimension'], y=df_knn['Accuracy(%)'], name = 'k-nearest neighbours',
                         line=dict(color='darkblue', width=2)))
fig.add_trace(go.Scatter(x=df_nb['Dimension'], y=df_nb['Accuracy(%)'], name='Naive Bayes',
                         line=dict(color='red', width=2)))
fig.add_trace(go.Scatter(x=df_mlp['Dimension'], y=df_mlp['Accuracy(%)'], name = 'multi-layer perceptron',
                         line=dict(color='maroon', width=2)))

fig.update_layout(title='Accuracy Measurement of Online Shoppers Purchasing Intention Dataset',
                  xaxis_title='Dimension',
                  yaxis_title='Accuracy (in percentage)',
                  xaxis = dict(dtick = 1),
                  yaxis=dict(dtick=1),
                  plot_bgcolor='rgb(220,220,235)')

Image(pio.to_image(fig, format='png'))
```
